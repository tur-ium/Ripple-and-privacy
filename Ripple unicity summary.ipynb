# Summary of the unicity of Ripple users the publicly available transaction history from 01/06/2017 - 29/08/2017

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results1 = pickle.load(open('results_p1_run2.pkl','rb'))
results2 = pickle.load(open('results_p2.pkl','rb'))
results3 = pickle.load(open('results_p3.pkl','rb'))
results4 = pickle.load(open('results_p4_run1.pkl','rb'))

plt.figure(figsize=(12,6))
font={'family': 'arial',
     'size': 24}
plt.title('Unicity of transactions for 01/06/2017 to 29/30/2017')
xpos = np.array(range(len(results1)))*2.5-.75
xpos2 = np.array(range(len(results1)))*2.5-.25
xpos3 = np.array(range(len(results1)))*2.5+.25
xpos4 = np.array(range(len(results1)))*2.5+.75
width = .5
plt.barh(xpos,list(results1.values())[::-1],tick_label=list(results1.keys())[::-1],height=width,label='1 point')
plt.barh(xpos2,list(results2.values())[::-1],tick_label=list(results2.keys())[::-1],height=width,label='2 points')
plt.barh(xpos3,list(results3.values())[::-1],tick_label=list(results3.keys())[::-1],height=width,label='3 points')
plt.barh(xpos4,list(results4.values())[::-1],tick_label=list(results4.keys())[::-1],height=width,label='4 points')
plt.legend(bbox_to_anchor=(-0.25,-.5,1.5,1),ncol=4,loc=3,mode='expand')
plt.ylabel('Coarsening')
plt.xlabel('Unicity')
```
